**Python Basics Question Answers**

1. What is Python, and why is it popular?
 -  Python is a high-level, interpreted programming language known for its readability and simplicity.
 -  It is popular because of its easy syntax, large community support, extensive libraries, and versatility in web development, data science, AI, automation, and more.
  - Example:-
     ```
     print("Hello, Python!")
     ```

2. What is an interpreter in Python?
- An interpreter is a program that executes Python code line by line, converting it into machine-readable code.
- Unlike a compiler, it does not produce a separate executable file.
- Example:-
  ```
  x = 5
  y = 3
  print(x + y)
  ```


3. What are pre-defined keywords in Python?
- Predefined keywords in Python are reserved words that have a special meaning and are used to define the syntax and structure of the language. Ex:- if, else, for, while, def, class, import, return.
- Example:-
  ```

  print(keyword.kwlist)
  ```

5. What is mutability in Python?
- Mutability in Python refers to whether an object’s value can be changed after it is created.
- Mutable objects can be modified, immutable objects cannot.
- Example:-
   ```
   my_list = [1, 2, 3] # Mutable
   my_list[0] = 10
   print(my_list)

   my_tuple = (1, 2, 3) # Immutable
   # my_tuple[0] = 10 # This would give an error
   ```

6. Why are lists mutable, but tuples are immutable?
- Lists are mutable because their elements can be changed, added, or removed.
- Tuples are immutable, meaning once created, their elements cannot be altered.
- Example:-
  ```
  list1 = [1, 2]
  list1.append(3)
  print(list1)

  tuple1 = (1, 2)
  # tuple1 += (3,) # Creates a new tuple
  ```

7. What is the difference between “==” and “is” operators?
- "==" checks if the values of two objects are equal.
- "is" checks if two objects refer to the same memory location.
- Example:-
   ```
   a = [1, 2]
   b = [1, 2]
   c = a
   print(a == b)
   print(a is b)
   print(a is c)
   ```
8. What are logical operators in Python?
- Logical operators in Python are used to combine conditional statements and return True or False. The three main logical operators are:-
1.  and → True if both conditions are True
2.  or → True if at least one condition is True
3.  not → Inverts the Boolean value
- Example:-
   ```
   x = 10
   y = 5
   print(x > 5 and y < 10)
   print(x > 5 or y > 10)
   print(not(x > 5))
   ```
9. What is type casting in Python?
- Type casting in Python is the process of converting a variable from one data type to another. Like:- int(), float(), str() etc.
- Example:-
   ```
   num_str = "10"
   num_int = int(num_str)
   print(num_int, type(num_int))
   ```

10. What is the difference between implicit and explicit type casting?  
- Implicit: Python automatically converts data type and the Implicit type casting also called as Type coercion.
- Explicit: Programmer manually converts type using functions like int(), float(), str(), and Explicit type casting also called as Type conversion.
- Example:-
   ```
   # Implicit
   a = 5
   b = 2.5
   c = a + b
   print(c, type(c))
   # Explicit
   d = float(a)
   print(d, type(d))
   ```
11. What is the purpose of conditional statements in Python?
- Conditional statements in Python are used to execute different blocks of code based on certain conditions.
- Example:-
   ```
   score = 65
   if score >= 50:
       print("Pass")
   else:
       print("Fail")
   ```
12. How does the elif statement work?
- The elif statement in Python is used to check multiple conditions after an if. It executes only when the if condition is false and its own condition is true. Only the true condition block runs, and the rest are skipped.  
- Example:-
   ```
   marks = 75
   if marks >= 90:
       print("Grade A")
   elif marks >= 70:
       print("Grade B")
   elif marks >= 50:
       print("Grade C")
   else:
       print("Fail")
   ```
13. What is the difference between for and while loops?
- For loop: A for loop is used to iterates over a sequence a definite number of times.
- While loop: A while loop is used when the number of interation is unknown and continues until a condition becomes false.
- For loop example:-
   ```
   for i in range(3):
       print("For loop iteration:", i)
   ```
- While loop example:-
   ```
   count = 0
   while count < 3:
       print("While loop iteration:", count)
       count += 1
   ```
14. Describe a scenario where a while loop is more suitable than a for loop?
- A while loop is more suitable when the number of interactions is not fixed and depends on a condition.
- Scenario:-
   ```
   Reading user input until they enter "exit". Here, we donot know how many times the user will type before exiting, so a while loop is better than a for loop.
   ```
- Example:-
   ```
   user_input = ""
   while user_input != "exit":
       user_input = input("Enter something (type 'exit' to stop): ")
       print("You entered:", user_input)
   ```
